```python
def find_item(cell):
    if pd.notna(cell):
        temp_list = set()
        for key, words_list in keywords_dict.items():
            for word in words_list:
                if word in cell.lower().replace('\xa0', '').split(','):
                    temp_list.add(key)
        if any(temp_list):
            return list(temp_list)
        else:
            return 'no matches'
    else:
        return cell


df['key'] = df['keywords'].apply(find_item)
```
