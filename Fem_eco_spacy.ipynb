```python
import pandas as pd
df=pd.read_excel('https://github.com/profrebrey/bibliometric_analysis/blob/main/tandfonline-02.xlsx?raw=true')
```

```python
df
```

```python
import spacy
```

```python
python -m spacy download en_core_web_sm
```

```python
nlp = spacy.load("en_core_web_sm")

def get_loc_with_spacy(text):
    result = []
    doc = nlp(text)
    for ent in doc.ents:
        if ent.label_ == 'ORG':
            result.append(ent.text)
    return result
```

```python
df['Author1_ORG']=df.Author1_Contacts.dropna().apply(get_loc_with_spacy)
df['Author2_ORG']=df.Author2_Contacts.dropna().apply(get_loc_with_spacy)
df['Author3_ORG']=df.Author3_Contacts.dropna().apply(get_loc_with_spacy)
df['Author4_ORG']=df.Author4_Contacts.dropna().apply(get_loc_with_spacy)
df['Author5_ORG']=df.Author5_Contacts.dropna().apply(get_loc_with_spacy)
df['Author6_ORG']=df.Author6_Contacts.dropna().apply(get_loc_with_spacy)
df['Author7_ORG']=df.Author7_Contacts.dropna().apply(get_loc_with_spacy)
df['Author8_ORG']=df.Author8_Contacts.dropna().apply(get_loc_with_spacy)
```

```python
df.to_csv('fem_eco.csv')
```
